```python
import os
from PIL import Image
import tkinter as tk
from tkinter import filedialog, messagebox

def resize_images(input_dir, output_dir, new_size):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Get the list of image files in the input directory
    image_files = [f for f in os.listdir(input_dir) if os.path.isfile(os.path.join(input_dir, f))]

    for file in image_files:
        # Open the image file
        image_path = os.path.join(input_dir, file)
        image = Image.open(image_path)

        # Resize the image
        resized_image = image.resize(new_size)

        # Save the resized image to the output directory
        output_path = os.path.join(output_dir, file)
        resized_image.save(output_path)

    messagebox.showinfo("Success", "Image resizing complete!")

def open_input_directory():
    input_directory = filedialog.askdirectory()
    input_dir_entry.delete(0, tk.END)
    input_dir_entry.insert(tk.END, input_directory)

def open_output_directory():
    output_directory = filedialog.askdirectory()
    output_dir_entry.delete(0, tk.END)
    output_dir_entry.insert(tk.END, output_directory)

def resize_images_gui():
    input_dir = input_dir_entry.get()
    output_dir = output_dir_entry.get()
    height = int(height_entry.get())
    width = int(width_entry.get())
    new_size = (height, width)

    if not os.path.isdir(input_dir):
        messagebox.showerror("Error", "Invalid input directory!")
        return

    if not os.path.isdir(output_dir):
        messagebox.showerror("Error", "Invalid output directory!")
        return

    resize_images(input_dir, output_dir, new_size)

# Create the main window
window = tk.Tk()
window.title("Image Resizer")

# Create labels and entry fields for input directory
input_dir_label = tk.Label(window, text="Input Directory:")
input_dir_label.grid(row=0, column=0, padx=10, pady=10, sticky="e")

input_dir_entry = tk.Entry(window, width=50)
input_dir_entry.grid(row=0, column=1, padx=10, pady=10)

input_dir_button = tk.Button(window, text="Browse", command=open_input_directory)
input_dir_button.grid(row=0, column=2, padx=10, pady=10)

# Create labels and entry fields for output directory
output_dir_label = tk.Label(window, text="Output Directory:")
output_dir_label.grid(row=1, column=0, padx=10, pady=10, sticky="e")

output_dir_entry = tk.Entry(window, width=50)
output_dir_entry.grid(row=1, column=1, padx=10, pady=10)

output_dir_button = tk.Button(window, text="Browse", command=open_output_directory)
output_dir_button.grid(row=1, column=2, padx=10, pady=10)

# Create labels and entry fields for image size
height_label = tk.Label(window, text="Image Height:")
height_label.grid(row=2, column=0, padx=10, pady=10, sticky="e")

height_entry = tk.Entry(window, width=10)
height_entry.grid(row=2, column=1, padx=10, pady=10)

width_label = tk.Label(window, text="Image Width:")
width_label.grid(row=2, column=2, padx=10, pady=10, sticky="e")

width_entry = tk.Entry(window, width=10)
width_entry.grid(row=2, column=3, padx=10, pady=10)

# Create the button to start image resizing
resize_button = tk.Button(window, text="Resize Images", command=resize_images_gui)
resize_button.grid(row=3, column=1, padx=10, pady=10)

# Start the main event loop
window.mainloop()
```
